## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]
y_df.info()
```

```python
# Create a list of at least 10 column names to use as X data
X_df = attrition_df[['Education','Age','DistanceFromHome','JobSatisfaction',
                     'OverTime','StockOptionLevel','WorkLifeBalance',
                     'YearsAtCompany','NumCompaniesWorked']]

# Show the data types for X_df
display(X_df.head())
display(X_df.info())
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary
X_train['OverTime'] = LabelEncoder().fit_transform(X_train['OverTime'])
X_test['OverTime']  = LabelEncoder().fit_transform(X_test['OverTime'])
X_train.head()
```

```python
# Create a StandardScaler, fit the train data, then scale the train and test data
X_train_scaled = StandardScaler().fit_transform(X_train)
X_test_scaled = StandardScaler().fit_transform(X_test)
```

```python
# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
department_encoder.fit(y_df[['Department']])

# Create two new variables by applying the encoder
# to the training and testing data
department_encoded_train = department_encoder.transform(y_train[['Department']])
department_encoded_test = department_encoder.transform(y_test[['Department']])

department_encoded_train[:5]
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False)

# Fit the encoder to the training data
attrition_encoder.fit(y_df[['Attrition']])

# Create two new variables by applying the encoder
# to the training and testing data
attrition_encoded_train = attrition_encoder.transform(y_train[['Attrition']])
attrition_encoded_test = attrition_encoder.transform(y_test[['Attrition']])

attrition_encoded_train[:5]
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
num_cols = X_train_scaled.shape[1]

# Create the input layer
input = layers.Input(shape=(num_cols,), name="input")

# Create the hidden layer
hidden1 = layers.Dense(units=64, activation='relu', name='hidden1')(input)

# Create the output layer
output1 = layers.Dense(units=2, activation='softmax', name='output1')(hidden1)

# Create at least two shared layers
shared1 = layers.Dense(units=64, activation='relu', name='shared1')(input)
shared2 = layers.Dense(units=128, activation='relu', name='shared2')(shared1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_hidden = layers.Dense(units=32, activation='relu', name='department_hidden')(shared2)

# Create the output layer
department_output = layers.Dense(units=3, activation='softmax', name='department_output')(department_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_hidden = layers.Dense(units=32, activation='relu', name='attrition_hidden')(shared2)

# Create the output layer
attrition_output = layers.Dense(units=2, activation='softmax', name='attrition_output')(attrition_hidden)
```

```python
# Create the model
model = Model(inputs=input, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'attrition_output':'binary_crossentropy',
                    'department_output':'categorical_crossentropy'},
              metrics={'attrition_output':'accuracy',
                    'department_output':'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_scaled,
    {
        'attrition_output': attrition_encoded_train,
        'department_output': department_encoded_train
    },
    epochs=50)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(
    X_test_scaled,
    {
        'attrition_output': attrition_encoded_test,
        'department_output': department_encoded_test
    })
```

```python
# Print the accuracy for both department and attrition
print(f"Department predictions accuracy: {results[3]}")
print(f"Attrition predictions accuracy: {results[4]}")
```
